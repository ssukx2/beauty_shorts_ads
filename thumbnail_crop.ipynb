### - 이미지 Crop 함수
** 예외처리: 에러발생 시 error 리스트에 이미지 이름 넣기

```python
error = []
path = ''
for videoId in df['videoId']:
    try:
        image_path = f'{path}{videoId}.png'
        output_path = f'{path}{videoId}.png'

        image = Image.open(image_path)
        width, height = image.size

        left = 437.5
        right = width - 437.5
        cropped_image = image.crop((left, 0, right, height))

        cropped_image.save(output_path)
                # 완료 메시지 출력
        print(f'{videoId}.png 저장 완료')

    except Exception as e:
        print(f'Error processing {videoId}: {e}')
        error.append(videoId)
```
